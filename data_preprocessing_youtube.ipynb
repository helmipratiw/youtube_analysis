input data

```python
import pandas as pd
data = pd.read_csv('dataset/trending.csv')
data.head()
```

menampilkan semua kolom

```python
data.columns
```

membuat data baru sesuai kebutuhan

```python
data = data[['title','channel_name','category_id','view','like','dislike','comment','trending_time']]
data.head()
```

membuat *category_id* jadi *category_name* dengan cara ambil data dari file .json

```python
import json

with open('dataset/category.json','rb') as f:
  categories = json.load(f)
```

```python
categories = pd.json_normalize(categories, record_path='items')
categories_dict ={idx:title for idx,
                  title in zip(categories['id'], categories['snippet.title'])}
```

sekarang mengubah *category_id* menjadi *category_names* lalu membuang *category_id*

```python
data['category'] = [categories_dict[str(x)] if str(x) in list(categories_dict.keys()) else 'unknown' for x in data['category_id']]
data
```

melihat semua kategori dan jumlah mereka

```python
data['category'].value_counts()
```

mengecek jika data masih ada yang tidak bernilai

```python
data.isna().sum()
```

mengubah semua nilai NA menjadi O

```python
data.fillna(0)
```

mengubah *trending_time* menjadi *trending_date*

```python
trending = pd.to_datetime(data['trending_time'], format='ISO8601')
data['trending_date'] = trending.dt.date
```

```python
data = data.drop(['trending_time'], axis=1)
```

melihat data duplikat

```python
data[['title','channel_name']].value_counts().reset_index(name='counts')
```

menghapus data duplikat

```python
used_data = data.drop_duplicates(['title','channel_name'], keep='first')
```

ubah menjadi pickle dan simpan

```python
used_data
```

```python
used_data.isna().sum()
```

```python
used_data = used_data.fillna(0)
```

```python
used_data.reset_index(drop=True)
used_data.drop('category_id', axis=1)
used_data.to_pickle('dataset/used_data.pickle')
```
